```python
import pandas as pd
from sklearn.datasets import load_iris
iris= load_iris()
iris
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.feature_names)
df
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names) # here we convert the all columns of iris dataset in dataframe
df
```

```python
df['target']=iris.target #here we add the new target column of in the data frame
df
```

```python
df['target'].nunique() # here we get the number of category are there which are unique
```

```python
iris.target_names[0],iris.target_names[1],iris.target_names[2] # here we get the names of the unique value/name of each category
```

```python
#here we make new column of flower names and in this we add the name of the unique category with the help of the target column
# ie. 0=setosa , 1= versicolor ,2= virginica with the help of lambda function
df['flower_name']=df.target.apply(lambda x: iris.target_names[x])
df
```

```python
# here we want to make 3 different frames with each of different category
df0=df[df.target==0] # same we can do for the other remaining 2 categories
df0
```

```python
# here we use slicing becoz we know frst 50 values belongs to category =0 and so on
df0=df[:50]
df1=df[50:100]
df2=df[100:]
df0
```

```python
# sepal length vs sepal width of setosa and versicolor
import matplotlib.pyplot as plt
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')


```

```python
# petal length vs petal width of setosa and versicolor
import matplotlib.pyplot as plt
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
```

```python
# making training and test sets
from sklearn.model_selection import train_test_split
x= df.drop(['target','flower_name'],axis='columns') #independent data
y= df.target  #dependent data
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
# create knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7) # by default n_neighbors is 5
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
# now we have to predict whether they belong to which class
knn.predict([[6.4,3.1,5.5,1.8]]) # it belongs to virginica as o/p=array([[2]])
```

```python
# now we have to build the cm matrix
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True) # annot is used to display numbers on color blocks also
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
# find the classification chart ie f-1 score, precision,recall
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# now check for which value of k the accuracy is more

for i in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    print("for K=",i,"accuracy is",knn.score(x_test,y_test))
```
